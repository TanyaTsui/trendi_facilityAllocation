# Reading and cropping rasters 
In this notebook, we convert the rasters downloaded from the [Annual Crop Inventory of Canada](https://open.canada.ca/data/en/dataset/ba2645d5-4458-414d-b196-6303ac06c1c9) into a shapefile of points, showing the ACI code at each location. Each ACI code represents a type of landuse for agriculture, such as 'greenhouses', 'oats', 'too wet to be seeded', 'wheat'...etc. Then, we ensure that the points are correctly categorized by province.

```python
import rasterio as rs
import numpy as np
from rasterio.plot import show
from rasterio.features import shapes
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import math
from fuzzywuzzy import fuzz, process
import time
pd.options.mode.chained_assignment = None  # default='warn'
```

## Read data 
* `crs_wkt` - crs (coordinate reference system) of Canada, in wkt (well known text) format. Used to unify crs of `canada` and `aci`
* `canada` - geodataframe of canada with province names and two letter province codes
* `aci` - geodataframe of aciCode for each location in Canada, created from raster files downloaded from the Annual Crop Inventory website. This is done in Arjang's google colab notebook `1_spatial_data_ingestion`, link [here](https://colab.research.google.com/drive/16FDURWrFnneBrh_45TjDzYka3zO8qTuT#scrollTo=pS7Tc4d5Jd56). Each ACI code represents a type of landuse for agriculture, such as 'greenhouses', 'oats', 'too wet to be seeded', 'wheat'...etc. 

cell below: `crs_wkt`

```python
crs_wkt = '''
PROJCRS["AAFC_Albers_Conical_Equal_Area",
    BASEGEOGCRS["WGS 84",
        DATUM["World Geodetic System 1984",
            ELLIPSOID["WGS 84",6378137,298.257223563,
                LENGTHUNIT["metre",1]],
            ID["EPSG",6326]],
        PRIMEM["Greenwich",0,
            ANGLEUNIT["Degree",0.0174532925199433]]],
    CONVERSION["unnamed",
        METHOD["Albers Equal Area",
            ID["EPSG",9822]],
        PARAMETER["Latitude of false origin",40,
            ANGLEUNIT["Degree",0.0174532925199433],
            ID["EPSG",8821]],
        PARAMETER["Longitude of false origin",-96,
            ANGLEUNIT["Degree",0.0174532925199433],
            ID["EPSG",8822]],
        PARAMETER["Latitude of 1st standard parallel",44.75,
            ANGLEUNIT["Degree",0.0174532925199433],
            ID["EPSG",8823]],
        PARAMETER["Latitude of 2nd standard parallel",55.75,
            ANGLEUNIT["Degree",0.0174532925199433],
            ID["EPSG",8824]],
        PARAMETER["Easting at false origin",0,
            LENGTHUNIT["metre",1],
            ID["EPSG",8826]],
        PARAMETER["Northing at false origin",0,
            LENGTHUNIT["metre",1],
            ID["EPSG",8827]]],
    CS[Cartesian,2],
        AXIS["(E)",east,
            ORDER[1],
            LENGTHUNIT["metre",1,
                ID["EPSG",9001]]],
        AXIS["(N)",north,
            ORDER[2],
            LENGTHUNIT["metre",1,
                ID["EPSG",9001]]]]
'''
```

```python
# read shpfile of Canada (with provinces), call it canada
canada = gpd.read_file('data/baseMapShp/canada_albers_long.shp')
canada = canada[canada.PRENAME.isin(['Alberta', 'British Columbia', 'Manitoba', 'New Brunswick', 'Newfoundland and Labrador', 
                                     'Nova Scotia', 'Ontario', 'Prince Edward Island', 'Quebec', 'Saskatchewan'])]

provinceCodes = {
    'Alberta': 'ab', 'British Columbia': 'bc', 'Manitoba': 'mb', 'New Brunswick': 'nb', 'Newfoundland and Labrador': 'nl', 
    'Nova Scotia': 'ns', 'Ontario': 'on', 'Prince Edward Island': 'pe', 'Quebec': 'qc', 'Saskatchewan': 'sk'
}

canada['province'] = canada.PRENAME.map(lambda x: provinceCodes[x])
canada = canada[['province', 'PRENAME', 'geometry']]
canada.to_file('data/baseMapShp/canada_albers_long.shp')
```

```python
%%time 
# extract geoms from rasters
shps = dict.fromkeys(['ab', 'bc', 'mb', 'nb', 'nl', 'ns', 'on', 'pe', 'qc', 'sk'])

for province in shps.keys(): 
    # print('extracting geoms from {} raster...'.format(province))
    mask = None
    with rs.Env():
        with rs.open('data/tif/3000m_aci_2020_{}_class.tif'.format(province)) as src:
            image = src.read(1) # first band
            results = (
            {'properties': {'raster_val': v}, 'geometry': s}
            for i, (s, v) 
            in enumerate(
                shapes(image, mask=mask, transform=src.transform)))
    geoms = list(results)

    # make shpfile, formatting
    gdf_fromGeom = gpd.GeoDataFrame.from_features(geoms)
    gdf_fromGeom.rename(columns={'raster_val': 'aciCode'}, inplace=True)
    gdf_fromGeom.aciCode = gdf_fromGeom.aciCode.astype('int64')
    gdf_fromGeom = gdf_fromGeom[gdf_fromGeom.aciCode > 0] # remove non-agricultural cells
    
    # add to shps dictionary 
    gdf_fromGeom['province'] = province
    shps[province] = gdf_fromGeom

aci = pd.concat(shps.values())
aci.geometry = aci.geometry.centroid
aci = aci.set_crs(crs_wkt)
```

## Categorise points by province 
The raster files provided by the Annual Crop Inventory (ACI) do not completely follow the province borders. In order to ensure that waste yield are correctly estimated in the next notebook `01_dataWrangling_2_calcYield`, we need to ensure that the points in the `aci` geodataframe are assigned to the correct province. 

```python
%%time 

# define function for filtering points from aci according to province 
def filterPts(province): 
    canada_p = canada[canada.province == province]
    aci_p = aci[aci.province == province]
    aci_p['withinP'] = aci_p.geometry.within(canada_p.iloc[0, -1])
    aci_p = aci_p[aci_p.withinP == True]   
    return aci_p

# filter points for each province 
provinces = ['ab', 'bc', 'mb', 'nb', 'nl', 'ns', 'on', 'pe', 'qc', 'sk']
toConcat = []

for province in provinces: 
    start = time.time()
    pts = filterPts(province)
    end = time.time()
    toConcat.append(pts)
    print('time taken for {}: {}s'.format(province, round(end-start)))
    
aci_all = pd.concat(toConcat)
```

## Format, save and visualise resulting geodataframe
The resulting geodataframe `aci_all` is a map of Canada that shows, at 3km resolution, the Annual Crop Inventory (ACI) code of each location. Each ACI code represents a type of landuse for agriculture, such as 'greenhouses', 'oats', 'too wet to be seeded', 'wheat'...etc.

```python
aci_all = aci_all[['province', 'aciCode', 'geometry']]
aci_all.to_file('data/shp/canada_aci.shp')
aci_all.head(3)
```

```python
fig, ax = plt.subplots(1,1,figsize=(30,30))
aci_all.plot(ax=ax, column='province', markersize=2)
canada.boundary.plot(ax=ax, color='black')
ax.axis('off')
plt.show()
```

## Limitations and further work
Limitations
* Rasters (geotiffs) from the annual crop inventory were downsampled from 800m to 3km resolution, to decrease file size and runtime. Because the resolution is lower, the locations of agricultural waste is less accurate. This is because some information is inevitably lost in the data aggregation process. 
* After downsampling, the rasters were then converted into points, to further decrease the file size and runtime. This further lowers the accuracy of the map of agricultural waste. 
* The resolution of 3x3km is acceptable for this country-wide analysis, because the size of Canada is around 4000 x 3000 km. In fact, when doing a country-wide analysis, the resolution could probably be even lower then 3km. 

Further work (optional, to improve accuracy of results) 
* We need to check how downsampling the rasters from 800m to 3km resolution affects the accuracy. Arjang has already checked this for 2km resolution. So we should either check the quality of the rasters at 3km resolution, or use 2km resolution rasters. 
* Perhaps even lower resolution rasters could be used, because this is a country-wide analysis. This depends on how much information is lost during the downsampling process.
